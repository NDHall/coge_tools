```python
import numpy as np 
def createSubPlots(dfa,ge,dfb):
    tes = dfa.columns.values.tolist()[3:]
    print(len(tes))
    num_rows = int( len(tes) /4 ) 
    print(num_rows)
    fig, ax = plt.subplots(num_rows,4,figsize=(35,55))
    i = 0
    j = 0
    for te in tes:
        if np.max(dfa[te]) > np.max(dfb[te]):
            maxX =  np.max(dfa[te])
        else:
            maxX =  np.max(dfb[te])
            
        if j == 4:
            j=0
            i += 1

        ax[i,j].set_title('{te}'.format(ge=ge,te=te))
        sns.kdeplot(dfa[te],color='red',ax=ax[i,j],)
        sns.kdeplot(dfb[te],color='blue',ax=ax[i,j])
        ax[i,j].set_xlim([-5,maxX])
        ax[i,j].set_ylim(0,2)
 
        
        j += 1
    plt.savefig('/home/ndh0004/code/coge_tools/images_notes/a_repetitive.pdf')
    #return num_rows
   
```

```python
print(createSubPlots(dfa,'A',dfb))
```
